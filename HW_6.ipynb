1. Вы открываете бутылку молока и делаете из нее глоток. Вероятность того, что молоко скисло - 0.02. С какой вероятностью молоко окажется свежим?

Вероятность обоих событий равна 1. Так как события зависимые, то есть произойдет либо одно, либо другое, то вероятность того, что при глотке молока не окажется признаков неудволетворительного состояния будет равна 1-0.02=0,98.
*Ответ: 0.98*

2. На пешеходном переходе установлен светофор с 2 лампами. Вероятность перегорания лампы в течение года - 0.2. С какой вероятностью не перегорит хотя бы одна лампа?

Вероятность перегорания каждой лампы - 0.2, соотвественно вероятность исправности - 0.8. Вероятность независимых событий равна их произведению - 0.64. *Ответ: 0.64*

3. Менеджер по продажам совершает сделку если знаком с клиентом с вероятностью - 0.8, а если ничего о нем не слышал с вероятностью 0.1. Начальник передает менеджеру список из 10 клиентов, 4 из которых менеджеру известны. С какой вероятностью выбрав случайного клиента из списка менеджер ничего ему не продаст?

Клиентов из каждой группы можно представить, как дробь от общего числа клиентов, то есть 4/10 - знакомых клиентов и 6/10 незнакомых клиентов. Если случайным образом клиент оказался знакомым, то сделка не пройдет с вероятностью - 0.2, а если незнакомый - 0.9. Значит, вероятность, что сделка не пройдет со знакомым человеком - произведение вероятности выпадения знакомого человека и вероятности, что сделка с ним не пройдет - 0.4 * 0.2 = 0.08, аналогичным образом рассчитаем с незнакомым человеком 0.6 * 0.9 = 0,54. Вероятность, которая будет характеризовать продажу случайно выбранному клиенту из списка, которому менеджер ничего не продаст, будет равна сумме вероятностей - 0.08 + 0.54 = 0.64.
